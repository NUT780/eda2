```python
#standard scaling - This technique scales the data so that the mean is 0 and the standard deviation is 1. 
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standard Scaling
scaler_standard = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']] = scaler_standard.fit_transform(df[['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[['age', 'education_num', 'capital_gain', 'capital_loss', 'hours-per_week']] = scaler_minmax.fit_transform(df[['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# One-Hot Encoding for categorical variables with less than 5 categories
one_hot_columns = ['sex', 'relationship', 'race']
df_one_hot_encoded = pd.get_dummies(df, columns=one_hot_columns)
```

```python
# Label Encoding for categorical variables with more than 5 categories
label_encoder = LabelEncoder()
label_columns = ['workclass', 'education', 'marital_status', 'occupation', 'native_country']
df_label_encoded = df.copy()
for col in label_columns:
    df_label_encoded[col] = label_encoder.fit_transform(df_label_encoded[col])
```

```python
# Create a new feature: age_bin
df['age_bin'] = pd.cut(df['age'], bins=[0, 25, 50, 75, 100], labels=['0-25', '26-50', '51-75', '76-100'])

# Create a new feature: capital_gain_loss (capital-gain - capital-loss)
df['capital_gain_loss'] = df['capital_gain'] - df['capital_loss']

# Display the new features
print(df[['age', 'age_bin', 'capital_gain', 'capital_loss', 'capital_gain_loss']].head())
```

```python
import numpy as np

# Apply log transformation to capital-gain
df['capital_gain_log'] = np.log1p(df['capital_gain'])

# Display the transformed feature
print(df[['capital_gain', 'capital_gain_log']].head())
```

```python
from sklearn.ensemble import IsolationForest

# Isolation Forest for outlier detection
isolation_forest = IsolationForest(contamination=0.1)
outliers = isolation_forest.fit_predict(df[['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])
df = df[outliers != -1]  # Remove outliers
```
